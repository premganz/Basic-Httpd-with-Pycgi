```python
# csv generator
import pandas as panda
import random
df = panda.DataFrame( columns=list('ABCDEF'))
for x in range(50000):
    a=random.randint(1,10)    
    b=random.randint(1,10)
    c=random.randint(1,10)
    d=random.randint(1,10)        
    e=random.randint(1,10)        
    if ((a>3 and b>3)  or (c>8)) and (e<5):
        f=1
    elif  ((a>7 and b>8) or (e>8)):
        f=2
    else:
        f=0
         
    df = df.append({'A':a,'B':b,'C':c,'D':d,'E':e, 'F':f}, ignore_index=True)
print(df[:50])
df.to_csv(r'C:/work/statis/cars/sum_train_large_v5.csv', index=False, header=False)
```
